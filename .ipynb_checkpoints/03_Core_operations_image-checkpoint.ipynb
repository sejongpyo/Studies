### Image Addition

```python
# difference between cv2 and np

x = np.uint8([250])
y = np.uint8([10])

print(cv2.add(x,y)) # 250+10 = 260 => 255

print(x+y)          # 250+10 = 260 % 256 = 4
```

### Image blending

- dst = imgA * a + imgB * b + c

```python
img1 = cv2.imread('/Users/sejongpyo/Downloads/yubi.jpg')
img2 = cv2.imread('/Users/sejongpyo/Downloads/(Blue).jpg')

dst = cv2.addWeighted(img1, 0.7, img2, 0.3, 0)

plt.imshow(dst)
plt.show()
```

```python
img1 = cv2.imread('/Users/sejongpyo/Downloads/yubi.jpg')
img2 = cv2.imread('/Users/sejongpyo/Downloads/(Blue).jpg')

rows, cols, channels = img2.shape
roi = img1[0:rows, 0:cols] # top left corner

img2gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
ret, mask = cv2.threshold(img2gray, 10, 255, cv2.THRESH_BINARY)
mask_inv = cv2.bitwise_not(mask)

# black-out the area
img1_bg = cv2.bitwise_and(roi, roi, mask = mask_inv)

# only region of logo from logo image
img2_fg = cv2.bitwise_and(img2, img2, mask = mask)

dst = cv2.add(img1_bg, img2_fg)
img1[0:rows, 0:cols] = dst

plt.imshow(img1)
plt.show()
```
