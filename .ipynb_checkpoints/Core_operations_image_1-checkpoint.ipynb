```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Access and modify pixel

```python
img = cv2.imread('/Users/sejongpyo/Downloads/yubi.jpg', )

px = img[100, 100] # row, column coordinates

print(px)
# array [Blue, Green, Red]
```

```python
# accessing only blue pixel
blue = img[100, 100, 0] # 1 = green, 2 = red

print(blue)
```

```python
# access RED value
img.item(100, 100, 2)
```

```python
# modify RED value
img.itemset((100, 100, 2), 100)
img.item(100, 100, 2)
```

### Image properties

```python
# rows, columns, channels
# grayscale image return (rows, columns)
print(img.shape)
```

```python
# size (total number of pixels)
print(img.size)

# datatype
print(img.dtype)
```

### Image ROI

```python
face = img[100:400, 200:430]
plt.imshow(cv2.cvtColor(face, cv2.COLOR_BGR2RGB))
plt.show
```

### Image Addition

```python
# difference between cv2 and np

x = np.uint8([250])
y = np.uint8([10])

print(cv2.add(x,y)) # 250+10 = 260 => 255

print(x+y)          # 250+10 = 260 % 256 = 4
```

### Image blending

```python
img1 = cv2.imread('/Users/sejongpyo/Downloads/yubi.jpg')
img2 = cv2.imread('/Users/sejongpyo/Downloads/(Blue).jpg')

img1.reshape(100, 100, 3)

dst = cv2.addWeighted(img1, 0.7, img2, 0.3, 0)

plt.imshow(dst)
plt.show()
```
