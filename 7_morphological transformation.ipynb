```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

- 단순화, 제거, 보정

- segmentate image and simplify, removal, adjustment it
- to check shape of image
- use binary or grayscale image  
    
- structuring element : kernel (square, ellipse, cross)

### erosion
- adapt structuring element and if there is '0', remove pixel
- effective for removal of tiny object


```python
img = cv2.imread('/Users/sejongpyo/downloads/gray.jpg')

kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))
dst = cv2.erode(img, kernel, iterations = 3)
# (src, kernel, anchor, iterations)
# anchor : structuring element 중심 default = (-1, -1)
# iterations : erosion 반복횟수


f = plt.figure(figsize = (18, 8))

ax1 = f.add_subplot(1, 2, 1)
ax1.imshow(img)
ax1.set_title('Original')

ax2 = f.add_subplot(1, 2, 2)
ax2.imshow(dst)
ax2.set_title('new')


plt.show()
```

### Dilation
- beside erosion, it fill a hole
- OR 연산 수행
- smoothen boundary and fill a hole


```python
img = cv2.imread('/Users/sejongpyo/downloads/gray.jpg')

kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))
dst = cv2.dilation(img, kernel, iterations = 3)
# (src, kernel, anchor, iterations)
# anchor : structuring element 중심 default = (-1, -1)
# iterations : dilation 반복횟수


f = plt.figure(figsize = (18, 8))

ax1 = f.add_subplot(1, 2, 1)
ax1.imshow(img)
ax1.set_title('Original')

ax2 = f.add_subplot(1, 2, 2)
ax2.imshow(dst)
ax2.set_title('new')


plt.show()
```
