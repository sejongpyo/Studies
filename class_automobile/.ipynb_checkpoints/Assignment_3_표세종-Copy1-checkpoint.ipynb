```python
import json
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import image
import numpy as np, copy

class sensorFusion:
    def __init__(self, cam0path, cam2lidarpath, imgpath, lidarpath, lblpath):
        self.cam0 = self.read_params(cam0path)
        self.cam2lidar = self.read_params(cam2lidarpath)
        self.img = self.read_img(imgpath)
        self.lidar = self.read_lidar(lidarpath)
        self.bbox = self.get_3dbbox(lblpath)
        
    def read_params(self, path):
        with open(path, 'r') as f:
            lines = f.read()
        li = []
        for i in lines.split('\n'):
            try:
                li.append(np.array([float(j) for j in i.split(' ')]))
            except:
                continue
        li = np.array(li)
        
        if li.shape[1] == 3:
            calib = np.eye(3)
        else:
            calib = np.eye(4)
        calib[:li.shape[0], :li.shape[1]] = li
        calib = np.asmatrix(calib)
        return calib
    
    def read_img(self, path):
        return image.imread(path)
    
    def read_lidar(self, path):
        return np.fromfile(path, dtype=np.float32).reshape((-1,5))
    
    def get_3dbbox(self, path):
        with open(path, 'r') as l:
            lidar_lbl = json.load(l)
        targets = []
        for i in lidar_lbl['annotations']:
            x = float(i['attribute']['location'][0])
            y = float(i['attribute']['location'][1])
            z = float(i['attribute']['location'][2])
            h = float(i['attribute']['dimension'][2])
            w = float(i['attribute']['dimension'][1])
            l = float(i['attribute']['dimension'][0])
            ry = i['attribute']['yaw']
            obj_lbl = np.array([x, y, z, l, w, h, ry])
            targets.append(obj_lbl)
        return targets
    
    def roty(self, t):
        c = np.cos(t)
        s = np.sin(t)
        return np.array([[c,  0,  s], [0,  1,  0], [-s, 0,  c]])
     
    def convert_3dbox_to_8corner(self, bbox3d_input):
        bbox3d = copy.copy(bbox3d_input)
        R = self.roty(bbox3d[6])
        l = bbox3d[3]
        w = bbox3d[4]
        h = bbox3d[5]

        x_corners = [l/2, l/2,  l/2, l/2,  -l/2, -l/2, -l/2, -l/2]
        y_corners = [-h/2, h/2,   h/2,  -h/2, -h/2, h/2, h/2, -h/2]
        z_corners = [-w/2, -w/2, w/2,  w/2,  -w/2, -w/2, w/2,  w/2]

        corners_3d = np.dot(R, np.vstack([x_corners,y_corners,z_corners]))

        corners_3d[0,:] = corners_3d[0,:] + bbox3d[0]
        corners_3d[1,:] = corners_3d[1,:] + bbox3d[1]
        corners_3d[2,:] = corners_3d[2,:] + bbox3d[2]

        return np.transpose(corners_3d)
    
    def point2img(self):
        points = self.lidar[:, 0:3]
        velo = np.insert(points, 3, 1, axis = 1).T
        cam = self.cam0 * self.cam2lidar[:3, :] * velo
        cam = np.delete(cam, np.where(cam[2, :] < 0)[1], axis = 1)
        
        cam[:2] /= cam[2, :]
        
        plt.figure(figsize=(12,5),dpi=96,tight_layout=True)
        png = self.img
        IMG_H,IMG_W,_ = png.shape

        plt.axis([0,IMG_W,IMG_H,0])
        plt.imshow(png)

        u,v,z = cam
        u_out = np.logical_or(u<0, u>IMG_W)
        v_out = np.logical_or(v<0, v>IMG_H)
        outlier = np.logical_or(u_out, v_out)
        cam = np.delete(cam,np.where(outlier),axis=1)

        u,v,z = cam
        plt.scatter([u],[v],c=[z],cmap='rainbow_r',alpha=0.5,s=2)
        plt.show()
        return cam
    
    def draw_bbox(self, cam, box_index):
        xyz = self.convert_3dbox_to_8corner(self.bbox[box_index])

        xyz1 = np.insert(xyz, 3, 1, axis = 1).T
        bbox = self.cam0 * self.cam2lidar[:3, :] * xyz1
        bbox[:2] /= bbox[2, :]
        
        x_min = bbox[0].min()
        y_min = bbox[1].min()
        x_max = bbox[0].max()
        y_max = bbox[1].max()

        plt.figure(figsize=(12,5),dpi=96,tight_layout=True)
        png = self.img
        IMG_H,IMG_W,_ = png.shape

        plt.axis([0,IMG_W,IMG_H,0])
        plt.imshow(png)

        u,v,z = cam
        u_out = np.logical_or(u<x_min, u>x_max)
        v_out = np.logical_or(v<y_min, v>y_max)
        outlier = np.logical_or(u_out, v_out)
        cam = np.delete(cam,np.where(outlier),axis=1)

        u,v,z = cam
        plt.scatter([u],[v],c=[z],cmap='rainbow_r',alpha=0.5,s=2)
        return plt.show()
    
    def get_dis2obj(self, box_index):
        xyz = np.expand_dims(self.bbox[box_index][:3], axis=0)
        xyz1 = np.insert(xyz, 3, 1, axis = 1).T
        xyz_cam =  self.cam2lidar[:3, :] * xyz1
        x, y, z = xyz_cam[0], xyz_cam[1], xyz_cam[2]
        dist = np.sqrt(x**2 + y**2 + z**2)
        
        return dist
    
    def draw_box_pts(self, index):
        corners_3d = self.convert_3dbox_to_8corner(self.bbox[0])
        xyz1 = np.insert(corners_3d, 3, 1, axis = 1).T
        bbox = self.cam0 * self.cam2lidar[:3, :] * xyz1
        bbox[:2] /= bbox[2, :]
        
        x = bbox[0][index]
        y = bbox[1][index]
        
        plt.figure(figsize=(12,5),dpi=96,tight_layout=True)
        png = self.img
        IMG_H,IMG_W,_ = png.shape

        plt.axis([0,IMG_W,IMG_H,0])
        plt.imshow(png)

        plt.scatter([x],[y])
        plt.show()
        return None
    
    def point2img_2(self):
        points = self.lidar[:, 0:3]
        velo = np.hstack((points, np.ones((points.shape[0], 1), dtype=np.float32)))
#         cam = velo * self.cam2lidar[:3, :] * self.cam0
        cam1 = np.dot(velo, self.cam2lidar[:3, :].T)
        cam = np.dot(cam1, self.cam0.T)
        print(cam.shape)
        cam = np.delete(cam, np.where(cam[2, :] < 0)[1], axis = 1)
        
        cam[:2] /= cam[2, :]
        
        plt.figure(figsize=(12,5),dpi=96,tight_layout=True)
        png = self.img
        IMG_H,IMG_W,_ = png.shape

        plt.axis([0,IMG_W,IMG_H,0])
        plt.imshow(png)

        u,v,z = cam
        u_out = np.logical_or(u<0, u>IMG_W)
        v_out = np.logical_or(v<0, v>IMG_H)
        outlier = np.logical_or(u_out, v_out)
        cam = np.delete(cam,np.where(outlier),axis=1)

        u,v,z = cam
        plt.scatter([u],[v],c=[z],cmap='rainbow_r',alpha=0.5,s=2)
        plt.show()
        return cam
```

## 라이다 포인트를 이미지 위에 겹쳐 그리기
### 과정
1. 라이다 데이터의 xyz에 해당되는 point들을 불러옴
2. 카메라 내부 파라미터, 외부 파라미터 (rotation, translation)를 불러온 point에 행렬곱 수행
3. 2D 이미지 상으로 투영된 모든 라이다 포인트에 거리 정보를 주기 위해 z축으로 0보다 작을 경우 해당 열 삭제
4. x, y축을 z축으로 나눔
5. 라이다 포인트가 이미지 밖의 범위에 위치한 열 삭제
6. depth에 대한 정보를 추가하여 라이다 포인트를 이미지 위에 시각화 실시

```python
sensorfusion = sensorFusion("data/calib_Camera0.txt", "data/calib_CameraToLidar0.txt",
                            'data/10008_084.jpg', 'data/10008_084.bin', 'data/10008_084.json')
cam = sensorfusion.point2img()
```

## 라벨링 (관심) 객체 라이다 포인트만 이미지 위에 겹쳐 그리기
### 과정
1. 주어진 객체에 대한 값인 x, y, z = 위치정보, h, w, l = 크기 정보, ry = 축회전 정보을 이용함. 3d bounding box의 가운데가 x, y, z이고 가로, 세로, 높이 정보를 활용하여 8개의 corners(8, 3)를 구함.
2. 각 점에 대해 앞서 수행한 방식대로 카메라 내, 외부 파라미터를 사용해 이미지 상으로 투영 실시
3. 3d bounding box에 대한 정보를 2d로 변환하기 위해 x, y 값에 대해 각각 가장 큰값과 작은값을 구함.
4. 앞서 투영한 라이다 포인트의 정보를 사용하여 2d bounding box보다 큰 값을 삭제함.

```python
def lidar_to_cam(boxes3d_lidar, lidar2cam):
    boxes3d_lidar_copy = copy.deepcopy(boxes3d_lidar)
    xyz_lidar = boxes3d_lidar_copy[0:3]
    l, w, h = boxes3d_lidar_copy[3], boxes3d_lidar_copy[4], boxes3d_lidar_copy[5]
    r = boxes3d_lidar_copy[6]

    xyz_lidar[2] -= h.reshape(-1) / 2
    xyz_lidar = xyz_lidar.reshape(-1, 3)
    pts_hom = np.hstack((xyz_lidar, np.ones((xyz_lidar.shape[0], 1), dtype=np.float32)))
    
    xyz_cam = np.dot(pts_hom, lidar2cam.T)
    r = -r - np.pi / 2
    return np.array([xyz_cam[0, 0], xyz_cam[0, 1], xyz_cam[0, 2], l, h, w, r], dtype=np.float32)
```
