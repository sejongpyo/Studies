# Meta data

```python
data = []

for i in train.columns:
    if i == 'OC':
        role = 'target'
    elif i == 'inst_id':
        role = 'id'
    else:
        role = 'input'
        
        
    if i == 'ownerChange' or i == 'OC':
        level = 'binary'
    elif i == 'sido' or i == 'instkind':
        level = 'categorical'
    elif i == 'inst_id':
        level = 'categorical'
    elif train[i].dtype == np.float64 and i != 'bedCount':
        level = 'interval'
    elif train[i].dtype == np.int64 or i == 'bedCount':
        level = 'ordinal'
        
    keep = True
    if i == 'inst_id':
        keep = False
        
    dtype = train[i].dtype
    
    i_dict = {
        'varname': i,
        'role' : role,
        'level': level,
        'keep':keep,
        'dtype':dtype
    }
    data.append(i_dict)
    
meta = pd.DataFrame(data, columns = i_dict.keys())
meta.set_index('varname', inplace = True)
```

# encoding for cat values

```python
v = meta[(meta.level == 'categorical') & (meta.keep)].index
train[v].apply(pd.Series.nunique, axis = 0)
```

```python
v = meta[(meta.level == 'categorical') & (meta.keep)].index
train = pd.get_dummies(train, columns = v, prefix = v, drop_first = True)
test = pd.get_dummies(test, columns = v, prefix = v, drop_first = True)
print('train: ', train.shape)
print('test: ', test.shape)
```

# encoding for binary values

```python
train.loc[(train['ownerChange'] == 'same'), 'ownerChange'] = int(0)
train.loc[(train['ownerChange'] == 'change'), 'ownerChange'] = int(1)
test.loc[(test['ownerChange'] == 'same'), 'ownerChange'] = int(0)
test.loc[(test['ownerChange'] == 'change'), 'ownerChange'] = int(1)
```

# encoding target values to numeric

```python
train.loc[(train['OC'] == 'open'), 'OC'] = 1
train.loc[(train['OC'] == ' close'), 'OC'] = 0
test.drop(labels = ['OC'], axis = 1, inplace = True)
```

```python
def change(x):
    if type(x) == str:
        if ',' in x:
            x = x.replace(',', '')
    return x

test['employee1'] = test['employee1'].apply(change)
test['employee2'] = test['employee2'].apply(change)
```

```python
for i in train.columns:
    if train[i].dtype == 'object':
        train[i]=train[i].astype(np.int64)
        
for j in test.columns:
    if test[j].dtype == 'object':
        test[j]=test[j].astype(np.int64)
```

```python
for i in train.columns:
    if i not in test.columns and i != 'OC':
        not_in_test = pd.DataFrame()
        not_in_test[i] = np.zeros((test.shape[0],), dtype=int)
        test = pd.concat([test, not_in_test], axis = 1)
        
for j in test.columns:
    if j not in train.columns:
        not_in_train = pd.DataFrame()
        not_in_train[j] = np.zeros((train.shape[0],), dtype = int)
        train = pd.concat([train, not_in_train], axis = 1)
```

```python
print('train:', train.shape)
print('test: ', test.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(train.drop(columns = ['OC', 'inst_id'], axis = 1), train['OC'], test_size = 0.3, random_state = 0, stratify = train['OC'])
```

# Randomforest

```python
rf_clf = RandomForestClassifier(random_state = 0, n_jobs = -1)
rf_clf.fit(X_train, y_train)
pred = rf_clf.predict(X_test)
accuracy = accuracy_score(y_test, pred)
print(round(accuracy * 100, 2))
```

# GridSearchCV

```python
params = {'n_estimators': [10, 100, 200, 300],
          'max_depth' : [6, 8, 10, 12],
          'min_samples_leaf' : [8, 12, 18],
          'max_features' : [30, 35]}

rf_clf = RandomForestClassifier(random_state = 0, n_jobs = -1)
grid_cv = GridSearchCV(rf_clf, param_grid = params, n_jobs = -1)
grid_cv.fit(X_train, y_train)

print('최적 하이퍼 파라미터: ', grid_cv.best_params_)
print('최고 예측 정확도: {:.4f}'.format(grid_cv.best_score_))
```

```python
rf_clf = RandomForestClassifier(max_depth = 6, max_features = 30, min_samples_leaf= 8, n_estimators = 10, random_state = 0, n_jobs = -1)
rf_clf.fit(X_train, y_train)
pred = rf_clf.predict(X_test)
accuracy = accuracy_score(y_test, pred)
print(round(accuracy * 100, 2))
```

```python
result_y = rf_clf.predict(test.drop(columns=['inst_id'], axis = 1))
```
