```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### theory

##### 동일한 이미지에 대해 사양한 사이즈가 필요할 때, 사용

- 가장 아래에 가장 큰 해상도를 놓고 점점 줄여가며 쌓아가는 형태

##### gaussian pyramid
- high level(낮은 해상도, 피라미드 상단), low level(높은 해상도, 피라미드 하단)
- 하단에서 row, column을 연속적으로 제거하며 생성

```python
img = cv2.imread('/Users/sejongpyo/downloads/yubi.jpg')

low = cv2.pyrDown(img) # 1/4 size
high = cv2.pyrUp(img) # 4 size

f, ax = plt.subplots(1, 3)
ax[0].imshow(img)
ax[1].imshow(low)
ax[2].imshow(high)
plt.show()
```

- pyrDown(), pyrUp()함수를 사용하면 계산 과정에서 약간의 차이에 의해 원본과 동일하지 않음

```python
# down -> up 한하면 반올림 등과 같은 변환으로 인해 원본 이미지와 1row차이를 지님.
# 원본 - resize(down -> up) = 외곽선

img = cv2.imread('/Users/sejongpyo/downloads/yubi.jpg')
print(img.shape) # (881, 630, 3)

GAD = cv2.pyrDown(img)
print(GAD.shape) # (441, 315, 3)

temp = cv2.resize(GAD, (630, 881))
res = cv2.subtract(img, temp)

plt.imshow(res)
plt.show()
```

## 이미지의 자연스러운 결합
1. gaussian pyramid 생성
2. 1번을 이용해 Laplacian pyramid 생성
3. Laplacian pyramid를 이용해 왼, 오 결합
4. 작은쪽을 확대하면서 동일 사이즈의 결합결과와 add하여 외곽선 선명하게 처리

```python
B = cv2.imread('/Users/sejongpyo/downloads/(Blue).jpg')
print(A.shape)
print(B.shape)
B.resize(881, 630, 3)
print(B.shape)
```

```python
A = cv2.imread('/Users/sejongpyo/downloads/yubi.jpg')
B = cv2.imread('/Users/sejongpyo/downloads/(Blue).jpg')
B.resize(881, 630, 3)
# 1
G = A.copy()
gpA = [G]
for i in range(6):
    G = cv2.pyrDown(G)
    gpA.append(G)

G = B.copy()
gpB = [G]
for i in range(6):
    G = cv2.pyrDown(G)
    gpB.append(G)

# 2
lpA = [gpA[5]]
for i in range(5, 0, -1):
    GE = cv2.pyrUp(gpA[i])
    temp = cv2.resize(gpA[i - 1], (GE.shape[:2][1], GE.shape[:2][0]))
    L = cv2.subtract(temp, GE)
    lpA.append(L)
    
lpB = [gpB[5]]
for i in range(5,0,-1):
    GE = cv2.pyrUp(gpB[i])
    temp = cv2.resize(gpB[i - 1], (GE.shape[:2][1], GE.shape[:2][0]))
    L = cv2.subtract(temp, GE)
    # L = cv2.subtract(gpB[i-1],GE)
    lpB.append(L)
    
# 3
LS = []
for la, lb in zip(lpA, lpB):
    rows, cols, dpt = la.shape
    ls = np.hstack((la[:, 0:cols//2], lb[:, cols//2:]))
    LS.append(ls)
    
# 4
ls_ = LS[0] # 좌측과 우측이 결합된 가장 작은 이미지
for i in range(1,6):
    ls_ = cv2.pyrUp(ls_) # Up Sacle
    temp = cv2.resize(LS[i],(ls_.shape[:2][1], ls_.shape[:2][0])) # 외곽선만 있는 이미지
    ls_ = cv2.add(ls_, temp) # UP Sacle된 이미지에 외곽선을 추가하여 선명한 이미지로 생성
    
real = np.hstack((A[:,:cols//2],B[:,cols//2:]))
f, ax = plt.subplots(1, 2)
ax[0].imshow(real)
ax[1].imshow(ls_)
```
