```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### theory
- searching and finding the location of a template image
- slides template image over original image
- return grayscale image
- output image size : (W - w + 1, H - h + 1)

```python
img = cv2.imread('/users/sejongpyo/downloads/airbus.jpeg', 0)
img2 = img.copy()
template = cv2.imread('/users/sejongpyo/downloads/bus.png', 0)
w, h = template.shape[::-1] # template.shape -> height, width [::-1] -> width, height

methods = ['cv2.TM_CCOEFF', 'cv2.TM_CCOEFF_NORMED', 'cv2.TM_CCORR',
            'cv2.TM_CCORR_NORMED', 'cv2.TM_SQDIFF', 'cv2.TM_SQDIFF_NORMED']

for meth in methods:
    img = img2.copy()
    method = eval(meth)
    
    # apply template matching
    res = cv2.matchTemplate(img, template, method)
    min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
    
    # TM_SQDIFF or TM_SQDIFF_NORMED -> take minimum
    if method in [cv2.TM_SQDIFF, cv2.TM_SQDIFF_NORMED]:
        top_left = min_loc
    else:
        top_left = max_loc
    bottom_right = (top_left[0] + w, top_left[1] + h)
    
    cv2.rectangle(img, top_left, bottom_right, 255, 2)
    
    plt.figure(figsize = (10,5))
    plt.subplot(121), plt.imshow(res, cmap = 'gray')
    plt.title('Matching result')
    plt.subplot(122), plt.imshow(img, cmap = 'gray')
    plt.suptitle(meth)
    
    plt.show()
```

## template matching multiple objects
- use thresholding

```python
img_rgb = cv2.imread('/users/sejongpyo/downloads/same.jpg')
img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2GRAY)
template = cv2.imread('/users/sejongpyo/downloads/jets.png', 0)
w, h = template.shape[::-1]

res = cv2.matchTemplate(img_gray, template, cv2.TM_CCOEFF_NORMED)
threshold = 0.3
loc = np.where(res >= threshold)
for pt in zip(*loc[::-1]):
    cv2.rectangle(img_rgb, pt, (pt[0] + w, pt[1] + h), (255, 0, 0), 50)
    
plt.imshow(img_rgb)
```
