```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

# fourier transform

#### 1. 개요
- 영상을 <U>주파수 성분</U>으로 변환하여 다양한 분석 및 처리
- 필터링 연산 fast fourier transform (FFT)를 이용하여 고속으로 구현 가능
- e.g. 섞여있는 물감을 하나씩 따로 분해하는것

#### 2. 이해
- <U>임의의 입력 신호를 다양한 주파수를 갖는 주기함수들의 합으로 분해하여 표현하는 것</U>
- 주파수 대역 - 분해 > sin, cos 주기함수

- red : input signal
- 파란색 모두를 합치면 red signal
- spatial domain -> frequency domain
- 입력 신호의 종류에 관계없이 모든 신호를 sin, cos 주기함수들의 합으로 분해 가능

#### 3. 수식

- 식 (1) : 입력신호 f(x)가 ej2πux의 합으로 표현 (푸리에 역변환)
- 식 (2) : f(x)를 주기함수 성분으로 분해했을 때 (푸리에 변환)


- e^j2piux : 주파수 u인 주기함수 성분 (e^2ipi == 한바퀴도는 속도) 2파이는 원의 둘레
- 허수를 사용하는 것은 평면에서 회전과 감기를 설명하는데 탁월함
- 관습적으로 시계 방향의 회전을 다루기 때문에 e승에 - 해줌

#### 4. 이미지에서 푸리에 변환
- 2차원의 신호



```python
# numpy

img = cv2.imread('/users/sejongpyo/downloads/mt.jpg', 0)
f = np.fft.fft2(img)
fshift = np.fft.fftshift(f) # 푸리에 변환 결과를 중심으로 이동
magnitude_spectrum = 20 * np.log(np.abs(fshift)) # 증폭 (변환값이 매우 작음)

plt.subplot(221), plt.imshow(img, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(magnitude_spectrum, cmap = 'gray')
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
img = cv2.imread('/users/sejongpyo/downloads/mt.jpg', 0)
f = np.fft.fft2(img)
fshift = np.fft.fftshift(f)

rows, cols = img.shape
crow,ccol = rows//2 , cols//2
fshift[crow-3:crow+3, ccol-3:ccol+3] = 0
f_ishift = np.fft.ifftshift(fshift)
img_back = np.fft.ifft2(f_ishift) # 역변환
img_back = np.real(img_back)

plt.subplot(131),plt.imshow(img, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(132),plt.imshow(img_back, cmap = 'gray')
plt.title('Image after HPF'), plt.xticks([]), plt.yticks([])
plt.subplot(133),plt.imshow(img_back)
plt.title('Result in JET'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# opencv

img = cv2.imread('/users/sejongpyo/downloads/mt.jpg', 0)
dft = cv2.dft(np.float32(img), flags = cv2.DFT_COMPLEX_OUTPUT)
dft_shift = np.fft.fftshift(dft)

magnitude_spectrum = 20*np.log(cv2.magnitude(dft_shift[:,:,0],dft_shift[:,:,1]))

plt.subplot(121),plt.imshow(img, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(magnitude_spectrum, cmap = 'gray')
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
img = cv2.imread('/users/sejongpyo/downloads/mt.jpg', 0)

dft = cv2.dft(np.float32(img),flags = cv2.DFT_COMPLEX_OUTPUT)
dft_shift = np.fft.fftshift(dft)

rows, cols = img.shape
crow,ccol = int(rows/2), int(cols/2)

# create a mask first, center square is 1, remaining all zeros
mask = np.zeros((rows,cols,2),np.uint8)
mask[crow-3:crow+3, ccol-3:ccol+3] = 1

# apply mask and inverse DFT
fshift = dft_shift*mask
f_ishift = np.fft.ifftshift(fshift)
img_back = cv2.idft(f_ishift)
img_back = cv2.magnitude(img_back[:,:,0],img_back[:,:,1])

plt.subplot(121),plt.imshow(img, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(img_back, cmap = 'gray')
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# simple averaging filter without scaling parameter
mean_filter = np.ones((3,3))

# creating a guassian filter
x = cv2.getGaussianKernel(5,10)
gaussian = x*x.T

# different edge detecting filters

# scharr in x-direction
scharr = np.array([[-3, 0, 3],
                   [-10,0,10],
                   [-3, 0, 3]])

# sobel in x direction
sobel_x= np.array([[-1, 0, 1],
                   [-2, 0, 2],
                   [-1, 0, 1]])

# sobel in y direction
sobel_y= np.array([[-1,-2,-1],
                   [0, 0, 0],
                   [1, 2, 1]])

# laplacian
laplacian=np.array([[0, 1, 0],
                    [1,-4, 1],
                    [0, 1, 0]])

filters = [mean_filter, gaussian, laplacian, sobel_x, sobel_y, scharr]
filter_name = ['mean_filter', 'gaussian','laplacian', 'sobel_x', \
                'sobel_y', 'scharr_x']

fft_filters = [np.fft.fft2(x) for x in filters]
fft_shift = [np.fft.fftshift(y) for y in fft_filters]
mag_spectrum = [np.log(np.abs(z)+1) for z in fft_shift]

for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(mag_spectrum[i],cmap = 'gray')
    plt.title(filter_name[i]), plt.xticks([]), plt.yticks([])
plt.show()
```
